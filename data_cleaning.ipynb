# Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('TMDB_movie_dataset_v11.csv')
df.head()
```

```python
# Convert to datetime format
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

start_date = '2014-01-01'
end_date = '2024-12-31'

# Filter data to only keep those within the last ~10 years
filtered_df = df[(df['release_date'] >= start_date) & (df['release_date'] <= end_date)]
filtered_df.head()
```

```python
# Filter to keep movies which are released
filtered_df = filtered_df[filtered_df['status'] == 'Released']
print(filtered_df.shape)
```

```python
# Remove data with 0 revenue
filtered_df = filtered_df[filtered_df['revenue'] > 100000]
print(filtered_df.shape)
```

```python
# Observe outliers
plt.figure(figsize=(10, 6))
plt.boxplot(filtered_df['revenue'])
plt.title("Revenue Box Plot")
plt.ylabel("Revenue")
plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = filtered_df['revenue'].quantile(0.25)
Q3 = filtered_df['revenue'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter for outliers
outliers = filtered_df[(filtered_df['revenue'] < lower_bound) | (filtered_df['revenue'] > upper_bound)]

print(outliers.shape)
```

```python
# Remove outliers based on above definition
filtered_df = filtered_df[(filtered_df['revenue'] >= lower_bound) & (filtered_df['revenue'] <= upper_bound)]
print(filtered_df.shape)
```

```python
# Observe difference with above boxplot after removing outliers
plt.figure(figsize=(10, 6))
plt.boxplot(filtered_df['revenue'])
plt.title("Revenue Box Plot")
plt.ylabel("Revenue")
plt.show()
```

```python
# Export cleaned data
filtered_df.to_csv('TMDB_movie_dataset_cleaned.csv', index=False)
```
